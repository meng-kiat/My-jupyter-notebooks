```python
import json

def extract_and_save_participant_data(matches, my_puuid, output_file, limit=201):
  """
  Extracts participant data for the given puuid from matches and saves it to a JSON file.

  Args:
    matches: A list of match dictionaries.
    my_puuid: The puuid of the player to extract data for.
    output_file: The path to the output JSON file.
    limit: The maximum number of matches to process.
  """

  participant_data = []
  found_matches = 0

  for match in matches:
    if found_matches >= limit:
      break

    for participant in match['info']['participants']:
      if participant['puuid'] == my_puuid:
        participant_data.append({
          # Add other desired data from the participant
          'participant': participant
        })
        found_matches += 1
        break

  with open(output_file, 'w') as f:
    json.dump(participant_data, f, indent=4)
```
